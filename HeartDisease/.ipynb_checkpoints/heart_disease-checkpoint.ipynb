```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Get the csv as a pandas dataframe
csv_path = './data/heart_disease_uci.csv'
df = pd.read_csv(csv_path)
# Show the 'head' of the dataset, default value 5
df.head()
```

```python
fig = plt.figure(figsize = (10, 5))

values_in_age = df['age'].value_counts()
age = []
freq_of_age = []
for age_, freq_of_age_ in values_in_age.items():
    age.append(age_)
    freq_of_age.append(freq_of_age_)

mean_age = np.mean(age)
bar_graph = plt.bar(age, freq_of_age)
plt.text(mean_age+1, max(freq_of_age)+.5, f'Mean Age: {mean_age:.2f}', color='green')
plt.axvline(x=mean_age, color='red')
plt.xlabel('Ages of People')
plt.ylabel('Freq of Age')
```

```python
df['num'] = [1 if num >= 1 else 0 for num in df['num']]
```

```python
df.head()
```

```python
values_in_num = df['num'].value_counts()
num_values = values_in_num.index
freq_of_num = values_in_num.values

# Create a bar plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(num_values, freq_of_num, color='skyblue')
ax.set_xlabel('Num (Heart Disease)')
ax.set_ylabel('Frequency of Heart Disease Cases')
ax.set_xticks([0, 1])
plt.show()
```

```python
target_df = df['num']
features_df = df.drop(['num', 'dataset', 'id'], axis=1)
target_df.head(), features_df.head()
```

```python
from sklearn.impute import SimpleImputer

# Define imputers for different strategies
median_imputer = SimpleImputer(strategy='median')
most_frequent_imputer = SimpleImputer(strategy='most_frequent')

# Apply median imputer for continuous numerical columns
features_df[['trestbps', 'chol', 'thalch', 'oldpeak']] = median_imputer.fit_transform(features_df[['trestbps', 'chol', 'thalch', 'oldpeak']])

# Apply most frequent imputer for categorical or binary columns
features_df[['fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']] = most_frequent_imputer.fit_transform(features_df[['fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']])

# Check for remaining missing values
print("Remaining missing values:\n", features_df.isnull().sum())
```

```python
df['cp'].value_counts()
```

```python
# We have some columns such as 'cp', 'slope', 'thalch', 'gender', 'restecg' that are all qualitative.
# Turn them to one-hot-encodings. This will allow us to represent them as as probalitity distribution. 
cp_encoded = pd.get_dummies(features_df['cp'], prefix='cp')
features_df = features_df.join(cp_encoded).drop('cp', axis=1)
features_df.head(2)
```

```python
# One-Hot Encode 'slope'
slope_encoded = pd.get_dummies(features_df['slope'], prefix='slope')
features_df = features_df.join(slope_encoded).drop('slope', axis=1)

# One-Hot Encode 'thal'
thal_encoded = pd.get_dummies(features_df['thal'], prefix='thal')
features_df = features_df.join(thal_encoded).drop('thal', axis=1)

# One-Hot Encode 'thalch'
thalch_encoded = pd.get_dummies(features_df['thalch'], prefix='thalch')
features_df = features_df.join(thalch_encoded).drop('thalch', axis=1)

# One-Hot Encode 'restecg'
restecg_encoded = pd.get_dummies(features_df['restecg'], prefix='restecg')
features_df = features_df.join(restecg_encoded).drop('restecg', axis=1)

# And who can forget, gender
features_df['sex'] = features_df['sex'].map({'Male': 0, 'Female': 1})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target_df, test_size=.1, random_state=42)
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()

# x_train, x_test = scaler.fit_transform(x_train), scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
print("accuracy:", accuracy)

# For Random Forest
f1_logreg = f1_score(y_test, y_pred)
print("Logreg F1 Score:", f1_logreg)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```

```python
# Accuracy for Random Forest
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)

# For Random Forest
f1_rf = f1_score(y_test, y_pred_rf)
print("Random Forest F1 Score:", f1_rf)
```
